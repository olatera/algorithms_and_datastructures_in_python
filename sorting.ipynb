## Сортировка в Python

### Алгоритм сортировки методом пузырька

Алгоритм сортировки «методом пузырька» основывается на попарном сравнении двух соседних элементов и, если следующий элемент массива оказывается меньше текущего, то значения элементов обмениваются местами, при этом наибольшее значение оказывается в конце массива («всплывший пузырёк»).


#### Описание алгоритма

1. Проход по неупорядоченному множеству (например, списку). При этом, если последующий элемент меньше предыдущего, то они
меняются местами. В итоге, самый «тяжелый» элемент оказывается в конце множества. 
2. Проход по множеству, исключая последний элемент, т.к. он уже отсортирован. При проходе обмен меньшего последующего на
больший предыдущий.
3. Количество проходов по множеству равно количеству элементов минус 1. Последний проход не нужен, т.к. остается один элемент, и
его значение меньше остальных.

```python
def bubble_sort(arrayToSort):
    a = arrayToSort
    for i in range(len(a),0,-1):
        for j in range(1, i):
            if a[j-1] > a[j]:
                tmp = a[j-1]
                a[j-1] = a[j]
                a[j] = tmp
    return a

ary = [11, 1, 0, 188, 19, 23, 5, 8, 2, 45, 69]
print (bubble_sort(ary))
```

### Cортировка выбором
#### Описание алгоритма

1 Найти наименьший элемент в массиве

2 Поменять местами его и первый элемент в массиве

3 Найти следующий наименьший элемент в массиве и поменять местами его и второй элемент массива

4 Продолжать это пока весь массив не будет отсортирован

```python
from IPython.display import YouTubeVideo
```

```python
def selection_sort(array):
    a = array
    for i in range(len(a)):
        idxMin = i
        for j in range(i+1, len(a)):
            if a[j] < a[idxMin]:
                idxMin = j
        tmp = a[idxMin]
        a[idxMin] = a[i]
        a[i] = tmp
    return a

ary = [11, 1, 0, 188, 19, 67, 886]
print (selection_sort(ary))
```

### Алгоритм сортировки вставками

 Суть его заключается в том что, на каждом шаге алгоритма мы берем один из элементов массива, находим позицию для вставки и вставляем. Стоит отметить что массив из 1-го элемента считается отсортированным.

```python
from IPython.display import YouTubeVideo
```

```python
YouTubeVideo('ROalU379l3U')
```

```python
def insertion_sort(array):
    a = array
    for i in range(len(a)):
        v = a[i]
        j = i
        while (a[j-1] > v) and (j > 0):
            a[j] = a[j-1]
            j = j - 1
        a[j] = v
    return a

ary = [54,1,2,3,52,3,1,2,3,543]
print (insertion_sort(ary))
```

### Быстрая сортировка(quicksort или qsort)

#### Описание алгоритма

На первом этапе необходимо выбрать из массива опорный элемент, относительно которого мы будем перемещать элементы. Для правильной работы алгоритма значение такого элемента должно лежать в интервале между минимальным и максимальным значениями элементов массива. Но зачастую берётся фиксированный элемент, например, средний или последний; можно также брать элемент со случайным индексом.
В нашей реализации мы будет использовать средний элемента.

Далее разбиваем исходный массив на три подмассива: в первом у нас будут храниться элементы со значением, которое меньше опорного элемента, во втором — элементы со значением равным опорному, а третий подмассив будет хранить значения большие, чем опорный.
После разделения рекурсивно упорядочиваем подмассивы, которые содержат значения находящиеся слева и справа от опорного
В итоге получаем отсортированную последовательность.

```python
def quick_sort(array=[12,4,5,6,7,3,1,15]):
    less = []
    equal = []
    greater = []

    if len(array) > 1:
        pivot = array[0]
        for x in array:
            if x < pivot:
                less.append(x)
            if x == pivot:
                equal.append(x)
            if x > pivot:
                greater.append(x)
        return quick_sort(less)+equal+quick_sort(greater)
    else: 
        return array

ary = [54,1,2,3,52,3,1,2,3,5,3,67,3,2,543]
print(quick_sort(ary))
```

#### Сортировка слиянием (англ. merge sort)

Сортировка слиянием(merge sort или mergesort) — алгоритм сортировки, основанный на сравнении элементов друг с другом. Эта сортировка является ярким примером использования парадигмы «разделяй и властвуй», была придумана Джоном фон Нейманом в 1945 году и опубликована в 1948 году.

1) Исходный массив рекурсивно разбивается на части, пока длина получающихся массивов не будет равна единице. Mассив единичной длины — отсортированный массив.
2) Отсортированные массивы соединяются по следующему принципу:
2.1) Сравниваются первые элементы уже отсортированных массивов. Меньший элемент помещается на первое место в результирующем массиве. Элементы массива, из которого был взят элемент, сдвигаются на один влево.
2.2) Как только один из массивов кончается, то элементы оставшегося массива помещаем в конец результирующего.

 В программе будет две функции: mergeSort() и merge(). Первая функция — основная, в ней список разбивается на части, а затем передаёт их в функцию merge(), которая сортирует и соединяет два списка в один.

```python
def merge_sort(inputList):
    if len(inputList) <= 1:
        return inputList

    middle = len(inputList) // 2
    left = inputList[:middle]
    right = inputList[middle:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)

def merge(left, right):
    res = []
    leftInd = 0
    rightInd = 0

    while leftInd < len(left) and rightInd < len(right):
        if left[leftInd] <= right[rightInd]:
            res.append(left[leftInd])
            leftInd += 1
        else:
            res.append(right[rightInd])
            rightInd += 1

    if left:
        res.extend(left[leftInd:])
    if right:
        res.extend(right[rightInd:])

    return list(res)

ary = [54,1,2,3,52,36,1,2,23,5,3,67,743,2,38,98,584]

print(merge_sort(ary))
```
